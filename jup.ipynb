```python
def trap_rain_water(height):
    # Guard: if the list is empty or has fewer than 3 bars, no water can be trapped
    if not height or len(height) < 3:
        return 0

    # Initialize two pointers at opposite ends of the elevation map.
    # The Two-Pointer technique lets us scan from both sides simultaneously,
    # avoiding the need to pre-compute left/right max arrays.
    left, right = 0, len(height) - 1

    # Track the highest bar seen so far from each side.
    # These act as the "water level ceiling" for each pointer's current position.
    left_max, right_max = 0, 0

    # Accumulator for total trapped water across all positions
    water_trapped = 0

    # --- TWO-POINTER APPROACH START ---
    # Concept: Instead of precomputing left_max[] and right_max[] arrays (O(n) space),
    # we use two inward-moving pointers. At any position, the trapped water is bounded
    # by min(left_max, right_max). We always move the pointer on the SMALLER side,
    # because that side is the limiting constraint — we already know its ceiling.
    # This gives O(n) time and O(1) space.

    while left < right:

        if height[left] < height[right]:
            # The left side is the limiting constraint right now.
            # The right side is definitely taller, so left_max is the water ceiling here.

            if height[left] >= left_max:
                # Current bar is taller than anything seen from the left so far.
                # No water trapped here — update the left ceiling instead.
                left_max = height[left]
            else:
                # Current bar is shorter than left_max, so water sits on top of it.
                # Water depth = left_max (ceiling) - height[left] (bar height)
                water_trapped += left_max - height[left]

            left += 1  # Move left pointer inward toward the center

        else:
            # The right side is the limiting constraint (height[right] <= height[left]).
            # right_max is the water ceiling for the right pointer's current position.

            if height[right] >= right_max:
                # Current bar is taller than anything seen from the right so far.
                # No water trapped here — update the right ceiling instead.
                right_max = height[right]
            else:
                # Current bar is shorter than right_max, so water sits on top of it.
                # Water depth = right_max (ceiling) - height[right] (bar height)
                water_trapped += right_max - height[right]

            right -= 1  # Move right pointer inward toward the center

    # --- TWO-POINTER APPROACH END ---

    return water_trapped


# Example usage:
if __name__ == "__main__":
    height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
    print(trap_rain_water(height))  # Output: 6
```
